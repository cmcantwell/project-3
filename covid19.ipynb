```python
insp = inspect(engine)
response = insp.get_table_names()
for table in response:
    print(f'{table} - - - - - - - - - - - - - - - - - - - - - -')
    cols = insp.get_columns(table)
    for col in cols:
        print(f'{col["name"]} - {col["type"]}')
```
